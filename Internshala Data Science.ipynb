# One Sample T-Test

```python
#importing libraries
import pandas as pd
import scipy.stats as stats
from scipy.stats import ttest_1samp
```

```python
#loading data
data = pd.read_csv("D:/Courses/Internshala Data Science/Module 3/one sample t-test/onesample.csv")
data = data['Insect Length']
```

```python
#printing first 5 rows
data.head()
```

Let's conduct a one sample t-test to check if the mean of the sample insects is similar to the mean of earlier insects.mean of earlier insects is. 6.09

We have the degree of freedom that is 28 and taking significance level to be 0.05, the t-critical value comes out to be 2.048

```python
t_statistic, _ = ttest_1samp(data, 6.09) #6.09 is population mean that is mean of insects 
```

```python
print(t_statistic)
```

t-statistic > t-critical
We reject the null Hypothesis.

# Paired T-test

```python
#Importing libraries
import pandas as pd
import scipy.stats as stats
from scipy.stats import ttest_rel
```

```python
#Loading data
data = pd.read_csv("D:/Courses/Internshala Data Science/Module 3/Paired t-test/Data for paired t test.csv")
```

```python
#Printing first 5 rows
data.head()
```

```python
# Calculating t and p-value using scipy library
t_statistic, p_value = stats.ttest_rel(data['Errors using typewriter'],data['Errors using a computer'])
```

```python
#printing p-value
p_value
```

As the p-value is less than 0.05, we can reject the null hypothesis (mean error from typewriter=mean error from computer)

```python
x_test=test.drop('Item_Outlet_Sales', axis=1)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lreg.fit(x_train,y_train)
```

```python
x_train=pd.get_dummies(x_train)
```

```python
x_train.head()
```

```python
x_test=pd.get_dummies(x_test)
```

```python
lreg.fit(x_train,y_train)
```

```python
#performance of our model -r2. Scikit learn provides a function score to check the performance of the model

lreg.score(x_test,true_p)
```

```python
lreg.score(x_train,y_train)
```

Here we can see our performance lowered/dropped from train to test

```python
#Evaluating the model using root mean square error (rmse) method

rmse_test = np.sqrt(np.mean(np.power((np.array(true_p)-np.array(pred_test)),2)))
rmse_train = np.sqrt(np.mean(np.power((np.array(y_train)-np.array(pred_train)),2)))
```

```python
print(rmse_train)
print(rmse_test)
```
